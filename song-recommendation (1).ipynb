```python
df = pd.read_csv('/kaggle/input/top-10000-spotify-songs-1960-now/top_10000_1960-now.csv')
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.head()
```

```python
df = df[['Track Name','Artist Name(s)','Album Name','Album Image URL','Track Preview URL','Artist Genres','Label']]
```

```python
df.head()
```

```python

df['Artist Name(s)1'] = df['Artist Name(s)'].str.replace(' ','')
```

```python
df['Album Name'] = df['Album Name'].str.replace(' ','')
df['Artist Genres'] = df['Artist Genres'].apply(lambda x: str(x).split(','))
```

```python
df['Label'] = df['Label'].str.replace(' ','')
```

```python
df['Artist Name(s)1'] = df['Artist Name(s)'].apply(lambda x: [x])
df['Album Name'] = df['Album Name'].apply(lambda x: [x])
df['Label'] = df['Label'].apply(lambda x: [x])
```

```python
df.head()
```

```python
df['tags'] = df['Artist Name(s)1']+df['Album Name']+df['Artist Genres']+df['Label']
```

```python
df.head()
```

```python
df = df[['Track Name','Album Image URL','Track Preview URL','tags','Artist Name(s)']]
```

```python
df.isna().sum()
```

```python
df.drop('Track Preview URL',axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df['tags'] = df['tags'].apply(lambda x:[str(i) for i in x])
df['tags'] = df['tags'].apply(lambda x: ' '.join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=9995, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi = cosine_similarity(vectors)
```

```python
simi[0]
```

```python
df[df['Artist Name(s)']=='Michael Jackson']
```

```python
def recommend(song):
    song_id = df[df['Track Name']==song].index[0]
    distances = simi[song_id]
    song_list = []
    song_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in song_list:
        print(df.iloc[i[0]]['Track Name'])
```

```python
recommend('Without Me')
```

```python
df.to_csv('final__df.csv')
```
